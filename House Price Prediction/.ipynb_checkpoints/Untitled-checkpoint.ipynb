## ðŸ”¸ Imports

```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error, f1_score
from sklearn.linear_model import Ridge
```

## ðŸ”¸ Load Dataset

```python
df = pd.read_csv('home_data.csv')
```

## ðŸ”¸ Preprocessing

```python
df.drop(['id', 'date', 'zipcode', 'lat', 'long'],axis=1,inplace=True)
```

```python
df['bathrooms'] = df['bathrooms'].astype(int)
df['floors'] = df['floors'].astype(int)
```

```python
df.rename(columns={'yr_renovated':'renovated'}, inplace=True)
```

```python
df['renovated'] = df['renovated'].apply(lambda x: 0 if x == 0 else 1)
```

```python
cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']
for col in cols:
    df[col] = df[col] / df[col].max()
```

## **ðŸ”¸ Exploratory Data Analysis (EDA)**

### Two Variable Analysis

```python
cols = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']
fig, axes = plt.subplots(2, 3, figsize=(10, 8))
for i, col in enumerate(cols):
    ax = axes[i//3, i%3]
    sns.scatterplot(x=col, y='price', data=df, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()    
```

```python
cols = ['renovated', 'floors', 'view', 'waterfront']
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

for i, col in enumerate(cols):
    ax = axes[i//2, i%2]
    sns.barplot(x=col, y='price', data=df, ax=ax)
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

## **ðŸ”¸ Evaluation**

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))

# Accuracy is not meaningful for regression; use R2 score instead
sns.barplot(
    x=['Gradient Boosting', 'Random Forest', 'Ridge'],
    y=[r2_score(y_test, gbr_pred), r2_score(y_test, rfr_pred), r2_score(y_test, rg_pred)],
    ax=ax[0, 0]
).set_title('R2 Score')

sns.barplot(
    x=['Gradient Boosting', 'Random Forest', 'Ridge'],
    y=[mean_squared_error(y_test, gbr_pred), mean_squared_error(y_test, rfr_pred), mean_squared_error(y_test, rg_pred)],
    ax=ax[0, 1]
).set_title('Mean Squared Error')

sns.barplot(
    x=['Gradient Boosting', 'Random Forest', 'Ridge'],
    y=[mean_absolute_error(y_test, gbr_pred), mean_absolute_error(y_test, rfr_pred), mean_absolute_error(y_test, rg_pred)],
    ax=ax[1, 0]
).set_title('Mean Absolute Error')

sns.barplot(
    x=['Gradient Boosting', 'Random Forest', 'Ridge'],
    y=[r2_score(y_test, gbr_pred), r2_score(y_test, rfr_pred), r2_score(y_test, rg_pred)],
    ax=ax[1, 1]
).set_title('R2 Score')

plt.tight_layout()
plt.show()
```
